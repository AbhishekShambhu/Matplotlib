```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=5,linestyle='dotted')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'rD-.')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'#111111',marker='D',linestyle='-.',markersize=15,alpha=0.2)
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')
plt.legend(loc='best',shadow=True,fontsize='large') #loc='upper right'
plt.grid()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95,color='g',histtype='step')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], 
         rwidth=0.95,color=['green','orange'],label=['men','women'],orientation='horizontal')
plt.legend()
```
